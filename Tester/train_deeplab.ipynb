## Testing

```python
#  This is needed because pytorch 2.6
filepath = './main.py'
old_line = "checkpoint = torch.load(opts.ckpt, map_location=torch.device('cpu'))"
new_line = "checkpoint = torch.load(opts.ckpt, map_location=torch.device('cpu'), weights_only=False)"

with open(filepath, 'r') as f:
    content = f.read()

content = content.replace(old_line, new_line)

with open(filepath, 'w') as f:
    f.write(content)
```
